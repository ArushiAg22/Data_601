```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
plot = sns.scatterplot(data= diamonds, x= "carat", y = "price", hue = "clarity", alpha = 0.5)

# Expectation was more the carats, higher the prize 
# but according to the plot, even diamonds with higher price range are of 1-3 carats mostly
```

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color")
```

#### 3.

- Use `carat` variable as as the size of the points in scatter plot in which x-axis is the `depth` and `price` on the y-axis.

```python
sns.relplot(data = diamonds, x="depth", y="price", size="carat", sizes=(20, 200), alpha=.5, height=6)
```
