```python
import pandas as pd
import re
```

```python
spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_spicy.csv'
less_spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_less_spicy.csv'
```

```python
df = pd.read_csv(less_spicy, index_col=0)
```

```python
df
```

```python
start = []
end = []
phone = []

for _, i in df["notes"].iteritems():
    s1 = i.split("end ")
    sdate = s1[0].replace("start ", "")
    start.append(sdate)
    pattern = re.compile('[Pp](#|\/n|n)(:)?( )+')      ## pattern check for phone numbers
    if (pattern.search(s1[1])):
        matched = pattern.search(s1[1])
        s2 = s1[1].split(matched.group())
        end.append(s2[0])
        phone.append(s2[1])
    else:
        end.append(s1[1])
        phone.append(None)    
```

```python
df["start_date"] = start
df["end_date"] = end
df["phone"] = phone
```

```python
df
```

```python
def remove_sp_char(df):
    """
    method to remove special chars (/ and -) from dates
    """
    for key, value in df.iteritems():
        pattern= re.compile('(\/|-)$')
        if (pattern.search(value)):
            mo=pattern.search(value)
            df[key]= value[:value.rfind(mo.group())]
```

```python
remove_sp_char(df.start_date)
remove_sp_char(df.end_date)
```

```python
df['start_date_new'] = pd.to_datetime(df['start_date'])
```

```python
df['end_date_new'] = pd.to_datetime(df['end_date'])
```

```python
df["duration_year"]= (df["end_date_new"].dt.year-df["start_date_new"].dt.year)
```
